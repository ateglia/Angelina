#3 - Numerical Calculator (if-elif-else statements)


```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Choose numerical operation (e.g., +, -, *, /, etc.): ")

res = None

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op == "//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation")

print(f"Result: {res}")
```

#9 - Dynamic dictionary; N distinct key:value pairs (while loop/ if-elif-else statements)

```python
N = int(input("Enter number of entries: "))

data = {}

i = 0

while i < N:
    name = input("Enter a name: ")

    if name in data:
        data[name] += 1
    else:
        data[name] = 1

    i += 1

print("Final dictionary:", data)
```

#12 - Recursive Function: String Reversal

```python
def reverse(s):
  if len(s) < 2:
    return s
  else:
    return s[-1] + reverse(s[:-1])

reverse('hello')
```

```python
def reverse(s):
  if len(s) < 2:
    return s
  else:
    #return s[-1] + reverse(s[:-1]) # Take the last char first
    #return reverse(s[1:]) + s[0] # Take the first char last
    return s[-1] + reverse(s[1:-1]) + s[0] # Swap first and last chars

reverse('12345')
```

#13 - ROI Calculator

```python
def roi(cost: float, revenue: float) -> float:
  assert cost > 0, "Cost must be positive"
  assert isinstance(cost, (int, float)), "Cost must be numerical"
  assert isinstance(revenue, (int,float)), "Revenue must be numerical"

  return(revenue - cost) / cost

if __name__ == "__main__":
  c = float(input('Enter cost: '))
  r = float(input('Enter revenue: '))
  ret = roi(c, r)
  print(ret)
```

#14 - Loan Payment Calculator

```python
def loan_payment(principal: float, annual_rate: float, months: int) -> float:
  assert isinstance(principal, (int,float)), "principal must be numeric"
  assert isinstance(annual_rate, (int,float)), "annual_rate must be numeric"
  assert isinstance(months, int), "months must be an integer"

  if principal <= 0:
    raise ValueError("Principal must be > 0.")
  if annual_rate < 0:
    raise ValueError("Annual rate cannot be negative.")
  if months <= 0:
    raise ValueError("Number of months must be > 0.")

  r = annual_rate / 12.0
  n = months
  P = float(principal)

  if r == 0.0:
    return P / n
  else:
    return r * P / (1.0 - (1.0 + r) ** (-n))

if __name__ == "__main__":
  try:
    P = float(input("Enter principal: "))
    apr = float(input("Enter annual interest rate as a decimal: "))
    n = int(input("Enter number of months: "))
    payment = loan_payment(P, apr, n)
    print(f"Monthly payment: {payment:.2f}")
  except Exception as e:
    print(e)
```

#15 - Quality Control Manager (exceptions)

```python
def getValues(n:int) -> list:
  '''
  Prompts the user for inputs, adds them to a lisst

  Input:
  - Number of items in the list

  Output:
  - A list
  '''
  items = []

  for i in range(n):
    try:
      item = float(input('Enter a weight value:'))
    except Exception as e:
      item = float('nan')

    items.append(item)
  print(items)
  return items
def calcAverage(alist: list) -> float:
  '''
  Input:
  - a list of weights

  Output:
  - the average weight
  '''
  try:
    return sum(alist[2:]) / len(alist[2:])
  except Exception as e:
    print(e)
    return float('nan')

def main():
  n = int(input('How many weights?'))
  if n < 0:
    raise ValueError('Number of items must be positive')

  L = getValues(n)
  avg = calcAverage(L)
  print(f'Average weight: {avg} lbs. ')

main()
```

#16 - Greeting Class

```python
class Greeting:
  def __init__(self, first, last):
    self.first = first
    self.last = last
  def __str__(self):
    return "Instance of class Greeting"

  def display(self, reverse=True):
    if reverse:
      print(f'{self.last}, {self.first}')
    else:
      print(self.first)
name = Greeting('John', 'Doe')
print(name)
name.display()
name.display(reverse=False)
```

#17 - Student Class

```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.__number = number
    self.__courses = []

  def get_name(self):
    return self.__name

  def get_courses(self):
    return self.__courses

  def set_name(self, newname):
    self.__name = newname

  def __len__(self):
    return len(self.__courses)

  def enroll(self, course):
    if course in self.__courses:
      print('You have already enrolled in', course)
    else:
      self.__courses.append(course)

s = Student('Bob', 12345)
s.get_name()
s.get_courses()
s.set_name('Bob Andrews')
s.get_name()
s.enroll('ABC123')
s.enroll('ABC123')
s.enroll('ABC567')
len(s)
s.get_courses()
```

```python
class Employee:
  def __init__(self, name, wage, hours):
    self.name = name
    self.__hourly_wage = wage
    self.hours = hours

  def get_wage(self):
    return self.__hourly_wage

  def set_wage(self, newWage):
    assert newWage > 0, "Wage values cannot be negative or zero."
    self.__hourly_wage = newWage

  def calculate_pay(self):
    return self.__hourly_wage * self.hours

  def __str__(self):
    return f"Employee: {self.name}\nHourly wages: {self.__hourly_wage}"

class Manager(Employee):
  def  __init__(self, name, wage, hours, bonus):
    super().__init__(name, wage, hours)
    self.bonus = bonus # assume it is given as float 0 -> 1 (e.g., 20% bonus is 0.2)

  def calculate_pay(self):
    return (1 + self.bonus) * self.get_wage() * self.hours # (1 + self.bonus) * super().calculate_pay()

employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)
print(employee)
print(manager)
print()
print(f"Employee {employee.name} weekly salary is {employee.calculate_pay()}$")
print(f"Manager {manager.name} weekly salary is {manager.calculate_pay()}$ that includes a {manager.bonus *100}% bonus")
```
