IN CLASS ASSIGNMENT #1

```python
sales = float(input("Enter sales ($):"))

cogs = 0.4 * sales
gross_profit = sales - cogs

print(f"Sales: {sales}$")
print(f"Cost of Goods Sold: {cogs}$")
print(f"Gross Profit: {gross_profit}$")
```

```python
sales = float(input("Enter sales ($): "))
tax_rate = float(input("Enter tax_rate (%)= "))

cogs = 0.4 * sales
gross_profit = sales - cogs
taxes = gross_profit * tax_rate / 100
net_income = gross_profit - taxes

print(f"Sales: {sales}$")
print(f"Cost of Goods Sold: {cogs}$")
print(f"Gross Profit: {gross_profit}$")
print(f"Tax Rate: {tax_rate}%")
print(f"Taxes: {taxes}$")
print(f"Net Income: {net_income}$")
```

IN CLASS ASSIGNMENT #2

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Choose numerical operation (e.g., +, -, *, /, etc.): ")

res = None

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op == "//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation")

print(f"Result: {res}")
```

```python
print("Hour\tFee")
for hour in range(1, 9):
  fee = 5 + 2.5 * hour
  if fee < 10.0:
    fee = 10.0
  elif fee > 20.0:
    fee = 20.0
  print(f"{hour}\t{fee}")
```

```python
print("Hour\tFee")
hour = 1
while hour < 9:
  fee = 5 + 2.5 * hour
  if fee < 10 :
    fee = 10
  elif fee > 20:
    fee = 20
  print(f"{hour}\t{fee}")
  hour += 1
```

IN CLASS ASSIGNMENT #4

```python
import math

ticketPrice = 10
initAttendees = 20

print("Ad$\tProfit")
for ad in range(0,201,25):
  revenue = (initAttendees + 2 *round(math.sqrt(ad))) * ticketPrice
  costs = 200 + ad
  profit = revenue - costs
  print(f"{float(ad)}\t{float(profit)}")
```

```python
import math

ticketPrice = 10
initAttendees = 20

print("Ad$\tProfit")
ad = 0
while ad < 201:
  revenue = (initAttendees + 2 *round(math.sqrt(ad))) * ticketPrice
  costs = 200 + ad
  profit = revenue - costs
  print(f"{float(ad)}\t{float(profit)}")
  ad += 25
```

IN CLASS ASSIGNMENT #5

```python
import random

check = input("Do you want to play a game (y/n)? ").lower()

while check == 'y':
  player = input("Choose b/w rock, paper scissors: ").lower()
  computer = random.choice(['rock', 'paper', 'scissors'])

  print(f"Player: {player} \nComputer: {computer}")

  if player == computer:
    print('Draw')
  elif player == 'rock' and computer == 'scissors':
    print('Player wins!')
  elif player == 'scissors' and computer == 'paper':
    print('Player wins!')
  elif player == 'paper' and computer == 'rock':
    print('Player wins!')
  else:
    print('Computer wins!')

  check = input("Do you want to play again (y/n)? ")
```

IN CLASS ASSIGNMENT #6

```python
numbers = []
num = int(input("Enter a number: "))

while num != 0:
  numbers.append(num)
  num = int(input("Enter another number: "))
numbers.sort(reverse=True)
print(numbers)
```

```python
numbers = []

while True:
  num = int(input("Enter a number: "))
  numbers.append(num)
  if num == 0:
    break

numbers = numbers [:-1]
numbers.sort(reverse=True)
print(numbers)
```

```python
weights = [10.0, 9.9, 10.1, 10.05]

s = 0
for w in weights:
  s += w

print(s/len(weights))

largest = 0
for w in weights:
  if w > largest:
    largest = w
print(largest)
```
