```python
def profit(money):
  import math
  ticketPrice = 10
  initAttendees = 20
  revenue = (initAttendees + 2 *round(math.sqrt(money))) * ticketPrice
  costs = 200 + money
  profit = revenue - costs
  print(f"{float(money)}\t{float(profit)}")
```

```python
print("Ad$\tProfit")
ad = 0
while ad < 201:
  profit(ad)
  ad += 25
```

```python
def quicksort(L):
  if len(L) < 2:
    return L
  else:
    pivot = L[len(L) // 2]
    less, equal, greater = [], [], []

    for item in L:
      if item < pivot:
        less.append(item)
      elif item == pivot:
        equal.append(item)
      else:
        greater.append(item)
  return quicksort(less) + equal + quicksort(greater)
quicksort([8,2,5,1,6,9,7])
```

```python
%timeit quicksort([8,2,5,1,6,9,7])
```

```python
class Car:
  def __init__(self, make, model, year, car_type, engine):
    self.make = make
    self.model = model
    self.year = year
    self.car_type = car_type
    self.engine = engine
    self.num_wheels = 4
    self.num_doors = 5
    self.engine_running = False
    self.speed = 0

  def engine_start(self):
    if self.engine_running:    # if self.engine_running == True
      print("The engine is running already...")
    else:
      self.engine_running = True

  def engine_stop(self):
    if not self.engine_running:  # if self.engine_running == False
      print("The engine is not running")
    else:
      self.engine_running = False

  def accelerate(self, acc, t):
    self.speed = min(self.speed + acc + t, 150)

  def decelerate(self, acc, t):
    self.speed = max(self.speed - acc + t, 0)
```

```python
v = Car("Tesla", "T3", 2026, "Sedan", "EV")
```

```python
v.speed
```

```python
v.engine_start()
v.accelerate(5, 3)
v.speed
```
